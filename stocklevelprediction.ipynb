The problem with the three datasets is the timestamp column because of the nature of its format. Hence, I proceed to convert this datatype into one that is easy to work with.

```python
def convert_to_datetime(data: pd.DataFrame = None, column: str = None):
  dummy = data.copy()
  dummy[column] = pd.to_datetime(dummy[column], format='%Y-%m-%d %H:%M:%S')
  return dummy
```

I have created a method in which we can load multiple objects, each containing a dataset, and specify the type of attribute to convert into a consistent format.

The function convert_to_datetime() takes a DataFrame and a column name as input and creates a copy of the DataFrame. Then it converts the specified column to datetime format using the format string '%Y-%m-%d %H:%M:%S' by calling the pd.to_datetime() function.

That is whether  or not it is possible to accurately predict the stock levels of products based on sales data and sensor data on an hourly basis in order to more intelligently procure products from their suppliers.

```python
from datetime import datetime

def convert_timestamp_to_hourly(data: pd.DataFrame = None, column: str = None):
  dummy = data.copy()
  newtest = dummy[column].tolist()
  newtest = [i.strftime('%Y-%m-%d %H:00:00') for i in newtest]
  newtest = [datetime.strptime(i, '%Y-%m-%d %H:00:00') for i in newtest]
  dummy[column]=newtest
  return dummy
```

The function starts by creating a copy of the input DataFrame using the copy() method.
It then converts the specified column of the DataFrame to a list of datetime objects using the tolist() method.
Next, it uses a list comprehension to convert each datetime object in the list to a string in the format '%Y-%m-%d %H:00:00' using the strftime() method.
After that, it uses another list comprehension to convert each string in the list back to a datetime object in the format '%Y-%m-%d %H:00:00' using the strptime() method from the datetime module.
Finally, it assigns the modified list back to the specified column of the DataFrame and returns the modified DataFrame.

```python
sls=convert_timestamp_to_hourly(sls, 'timestamp')
sls.head()
```

We now set y as our dependent variable and x as our independent variable


```python
x=merged_ds.drop(columns=['estimated_stock_pct'])
y=merged_ds['estimated_stock_pct']
print(x.shape)
print(y.shape)
```

```python
k=10
split=0.75

```

the supervised ML algorithms we are going employ to train the ML model are tree-based, regression, and classification because we have categorized and defined all the inputs in the database and know what to look for in the output.
```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
```

We now create a loop to train K models with a 75/25% random split of the data each time between training and test samples

```python
accuracy=[]

for fold in range(0, k):
  #we instantiate the algorithm
  model=RandomForestRegressor()
  scaler=StandardScaler()

  #we create training and test samples
  Xtrain, Xtest, Ytrain, Ytest = train_test_split(x, y, train_size=split, random_state=42)

  #we scale X data because it helps the algorithm to converge and not take in large values
  scaler.fit(Xtrain)
  Xtrain=scaler.transform(Xtrain)
  Xtest=scaler.transform(Xtest)

  #Train model
  trained_model=model.fit(Xtrain, Ytrain)

  #Generate predictions on test sample
  y_pred = trained_model.predict(Xtest)

  #Compute accuracy using mean absolute error
  mae = mean_absolute_error(y_true=Ytest, y_pred =y_pred)
  accuracy.append(mae)
  print(f"Fold {fold + 1}: MAE = {mae:.3f}")

print(f"AverageMAE: {(sum(accuracy)/len(accuracy)):.2f}")



```

If the mean absolute error (MAE) is almost exactly the same each time, it is a good sign because it shows that the performance of the model is consistent across different random samples of the data.

We chose the MAE as the performance metric because it describes how closely the machine learning model was able to predict the exact value of the estimate_stock_pct

Even though the model is consistent, the MAE's value is not good enough. Since the average value of the target variable is around .51, meaning that the accuracy as a percentage was around 50%. We would want the MAE to be as low as possible. This is where the iterative process of ML comes in. At this stage, since we only have small samples of the data, we can report back to the business within these findings and recommend that the dataset needs to be further engineered, or more datasets need to be added.

As a final note, we can use the trained model to interpret which features were significant when the model was predicting the target variable. We will use matplotlib and numpy to visualize the results, wo we should install and import this package.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
features=[i.split("__")[0] for i in x.columns]
importances = model.feature_importances_
indices = np.argsort(importances)

fig, ax = plt.subplots(figsize=(10, 20))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

temperature, hour of the day, unit price were important in predicting stock levels. However, product categories and other features were not as relevant.
