```python
df[['customer_type']].value_counts()
```

Running the method count values for the feature (attribute) customer_type, we learn how many observations fall into the categories non-member, standard, premium, basic, and gold. The largest type of consumers at Gala Groceries are non-members, followed by standard, and premium.

```python
df[['customer_type']].value_counts(normalize=True)
```

```python
df[['category']].value_counts()
```

Repeting the same study for the category attribute. Fruit and vegetables are the two most frequently purchased items at Gala Groceries.

```python
df[['category']].value_counts(normalize=True)
```

Respresenting the frequency of purchases by category using percentages.

```python
df.groupby('category')[['unit_price']].mean()
```

Presenting the average unit cost per category at the store.

```python
df.groupby('category')[['unit_price']].median()
```

Presenting the median cost per category at the store after revising the nature of the sample data.

```python
df.groupby('category')[['quantity']].mean().sort_values(by='quantity')
```

Sorting the quantities of items per category at the store from the lowest to the highest using the mean measurement.

```python
df.groupby('category')[['quantity']].median().sort_values(by='quantity')
```

Sorting the volume of items per category from the lowest to the highest using the median measurement.

```python
df.groupby('category')[['quantity']].agg(['mean', 'median', 'min', 'max'])
```

This line of code outputs a summary statistics for items' volume per category by mean, minimum, and maximum values. Clearly, the mean values are not optimal for our study because the are eveident outliers in the dataset.
