```python
df['total'].plot(kind='hist')
```

The histogram shows that purchases less than or equal to $20 are the most frequent at the store. Thus, the sample data is skewed toward small purchases per transaction. Because the data is not normally distributed, we resort to median as a more accurate representation of expected results per event.

```python
df.pivot(columns='category', values='quantity')
```

Running a pivot table to represent the values of the volume per category at the store, the code returns a table with many NaNs, or missing values. Hence, I move on to another graphical representation of the data.

```python
df.groupby('timestamp')['category'].value_counts().unstack()
```

The line of code above produces an expandable table that displays the values per category purchased at every recorded timestamp.
